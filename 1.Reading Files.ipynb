```python
#No. of words in the file
#word : a set of charecters ending with a space
with open('MyTextFile1.txt','r') as f:
    count=0
    for line in f:
        #removes all the leading and trailing non printable charecters
        line = line.strip()
        line=line.split(" ")
        count+=len(line)
print(count)
```

```python
#Understanding how split and strip works
s = "  removes all the leading and trailing non printable charecters \n "
s=s.strip()
print(s)
line = s.split(" ")
print(line)
print(len(line))
```

```python
with open('MyTextFile1.txt','r') as f:
    words =[]
    for line in f:
        line = line.strip()
        line=line.split(" ")
        #Preparing a list of all the words accross every line
        words.extend(line)
print(words[:20])
print(len(words))
print(len(set(words)))
```
