```python
nlp = en_core_web_sm.load()
```

For the Input text you will be able to fetch:
 1. POS tags
 2. Entities
 3. Dependencies

### Parts of the Speech  - Class of words based on the word's function, the way it works in a sentence. The parts of speech are noun, verb, adjective, adverb, pronoun, preposition, conjunction, and interjection.


```python
res = []
```

```python
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")

for token in doc:
    res.append([token.text, token.lemma_, token.pos_, token.tag_, token.is_alpha, token.is_stop])
```

```python
pd.DataFrame(res,columns=["text","lemma_","pos","tag","is_alpha","is_stop"])
```

### Entities - Anything thing which is distinct and has an independent existence


```python
doc = nlp('India authorities fined google a record $5.1 billion \
          on Monday for some bad practice')
```

```python
print([(X.text, X.label_) for X in doc.ents])
```

### Understanding the Dependencies
- Every token has a garmmatical relationship between other tokens which are bounded by the grammer defined by the language
- These relationships are Weaved together to form a sentence 
- To generate the grammatical relations between the tokens, we use linguistic parsers and syntactic dependency parsing is one of them. 
- Via dependency parsing, we create a tree or a graph data structure of a sentence conveying its tokens' grammatical relations. 
- Parsers generally performs the below tasks for an input sentence:
    - tokenize
    - tags with POS 
    - Construct the dependency trees

```python
doc1 = nlp('where is Anil Working')
```

```python
doc2 = nlp('Anil is working with google')
```

```python
displacy.render(nlp(str(doc1)),style = 'dep')
```
